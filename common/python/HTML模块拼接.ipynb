```python
import os
import urllib.request
import webbrowser

def start(htmldir):
    global indexdir    
    if not os.path.exists(htmldir):
        os.mkdir(htmldir)        
    indexdir = os.path.join( htmldir, "index")
    indexdir = urllib.request.url2pathname(indexdir)
    if not os.path.exists(indexdir):
        os.mkdir(indexdir)    
        
def txt_link(indexdir, htmldir):
    opentxt = os.path.join( indexdir, "link.txt")
    opentxt = urllib.request.url2pathname(opentxt)
    rel_ico = os.path.relpath("../image/favicon.ico", htmldir)
    rel_ico = urllib.request.pathname2url(rel_ico)
    rel_css = os.path.relpath("../css/mystyle.css", htmldir)
    rel_css = urllib.request.pathname2url(rel_css)
    message = """
  <link href="%s" rel="shortcut icon" type="image/x-icon" />
  <!--调用网页图标-->
  <link href="%s" rel="stylesheet" type="text/css" />
  <!--调用css样式-->
    """%(rel_ico, rel_css)
    fw = open(opentxt, 'w', encoding='utf-8')
    fw.write(message)
    fw.close()    
    
def txt_title(indexdir, htmldir):
    opentxt = os.path.join( indexdir, "title.txt")
    opentxt = urllib.request.url2pathname(opentxt)
    message = """
  <title> | 颐亲王的小栈</title>
  <meta name="keywords" content="颐亲王, 郑協, zhengxie.info, 正协, 正协信息" />
  <meta name="description" content="本站点是“颐亲王的小栈”，站点名称为“正协信息客栈”，由“华朝颐亲王郑協”创建并维护，借助“GitHub Pages”，发布于域名“zhengxie.info”内。" />
    """
    if not os.path.exists(opentxt):
        fw = open(opentxt, 'w', encoding='utf-8')
        fw.write(message)
        fw.close()
        
def txt_author(indexdir, htmldir):
    opentxt = os.path.join( indexdir, "author.txt")
    opentxt = urllib.request.url2pathname(opentxt)
    message = """
  <meta name="author" content="华朝颐亲王郑協, seymour.zx@foxmail.com" />
   """
    if not os.path.exists(opentxt):
        fw = open(opentxt, 'w', encoding='utf-8')
        fw.write(message)
        fw.close()

def txt_article(indexdir, htmldir):
    opentxt = os.path.join(indexdir, "article.txt")
    opentxt = urllib.request.url2pathname(opentxt)
    message = """
        <h1>欢迎作客颐亲王的小栈</h1>
        <p></p>
        <br /><br /><br /><br /><br />
        <br /><br /><br /><br /><br />
        <br /><br /><br /><br /><br />
        <br /><br /><br /><br /><br />
        <br /><br /><br /><br /><br />
    """
    if not os.path.exists(opentxt):
        fw = open(opentxt, 'w', encoding='utf-8')
        fw.write(message)
        fw.close()
    
def create_html(htmldir):
    openhtml =  os.path.join(htmldir, "index.html")
    openhtml = urllib.request.url2pathname(openhtml)
    list_txt = []

    txtpath = os.path.abspath("../txt/head.txt")
    list_txt.append(txtpath)

    txtpath = os.path.join(os.path.abspath(indexdir), "title.txt")    
    list_txt.append(txtpath)

    txtpath = os.path.join(os.path.abspath(indexdir), "link.txt")    
    list_txt.append(txtpath)

    txtpath = os.path.join(os.path.abspath(indexdir), "author.txt")    
    list_txt.append(txtpath)

    txtpath = os.path.abspath("../txt/body.txt")
    list_txt.append(txtpath)

    txtpath = os.path.abspath("../txt/header.txt")
    list_txt.append(txtpath)

    txtpath = os.path.abspath("../txt/centrals.txt")
    list_txt.append(txtpath)

    txtpath = os.path.abspath("../txt/mains.txt")
    list_txt.append(txtpath)

    txtpath = os.path.abspath("../txt/articles.txt")
    list_txt.append(txtpath)

    txtpath = os.path.join(os.path.abspath(indexdir), "article.txt")    
    list_txt.append(txtpath)

    txtpath = os.path.abspath("../txt/articlee.txt")
    list_txt.append(txtpath)
    
    txtpath = os.path.abspath("../txt/maine.txt")
    list_txt.append(txtpath)
    
    txtpath = os.path.abspath("../txt/centrale.txt")
    list_txt.append(txtpath)

    txtpath = os.path.abspath("../txt/footer.txt")
    list_txt.append(txtpath)
    
    fw = open(openhtml, 'w', encoding='utf-8')
    for i in range(len(list_txt)):
        if not os.path.exists(list_txt[i]):
            print("文件",list_txt[i],"不存在")
        else:            
            with open(list_txt[i], 'r', encoding='utf-8') as fr:
                fw.write('\n')
                for j in fr:
                    fw.write(j)
                fr.close()
    fw.close()
    webbrowser.open(openhtml,new = 0, autoraise=True)

def run(pydir, htmldir):
#     程序开始，初始化    
    start(htmldir=htmldir)    
    txt_link(indexdir=indexdir, htmldir=htmldir)
    txt_title(indexdir=indexdir, htmldir=htmldir)
    txt_author(indexdir=indexdir, htmldir=htmldir)
    txt_article(indexdir=indexdir, htmldir=htmldir)
    create_html(htmldir=htmldir)

# 为了防止再次运行时位置出错，记录py所在目录，最后要返回初始位置。
pydir = "D:\\Workshop\\html\\seymour-zx.github.io\\common\\python"
os.chdir(pydir)    
print('py所在目录：pydir =', os.getcwd())

list_htmldir = ["../../base/test", "../../base/about", "../../base/changelog", "../../base/copyright", "../../base/disclaimer", "../../base/homepage", "../../base/privacy", "../../base/sitemap", "../../private/bookmark"]
# for n in range(2):
#     newhtmldir = "../../base/test" + "/" + str(n+1)
#     list_htmldir.append(newhtmldir)

for nhtml in range(len(list_htmldir)):
    run(pydir=pydir, htmldir=list_htmldir[nhtml])
```
