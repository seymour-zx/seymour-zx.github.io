```python
import os
import urllib.request
import webbrowser
import time

def txt_link(indexdir, htmldir):
    opentxt = os.path.join( indexdir, "link.txt")
    opentxt = urllib.request.url2pathname(opentxt)
    rel_ico = os.path.relpath("../image/favicon.ico", htmldir)
    rel_ico = urllib.request.pathname2url(rel_ico)
    rel_css = os.path.relpath("../css/mystyle.css", htmldir)
    rel_css = urllib.request.pathname2url(rel_css)
    message = """
  <link href="%s" rel="shortcut icon" type="image/x-icon" />
  <!--调用网页图标-->
  <link href="%s" rel="stylesheet" type="text/css" />
  <!--调用css样式-->
    """%(rel_ico, rel_css)
    fw = open(opentxt, 'w', encoding='utf-8')
    fw.write(message)
    fw.close()    
    
def txt_author(indexdir, htmldir):
    opentxt = os.path.join( indexdir, "author.txt")
    opentxt = urllib.request.url2pathname(opentxt)
    message = """
  <meta name="author" content="华朝颐亲王郑協, seymour.zx@foxmail.com" />
   """
    if not os.path.exists(opentxt):
        fw = open(opentxt, 'w', encoding='utf-8')
        fw.write(message)
        fw.close()

def txt_article(indexdir, htmldir):
    opentxt = os.path.join(indexdir, "article.txt")
    opentxt = urllib.request.url2pathname(opentxt)
    message = """
        <h1>欢迎作客颐亲王的小栈</h1>
        <p></p>
        <br /><br /><br /><br /><br />
        <br /><br /><br /><br /><br />
        <br /><br /><br /><br /><br />
        <br /><br /><br /><br /><br />
        <br /><br /><br /><br /><br />
    """
    if not os.path.exists(opentxt):
        fw = open(opentxt, 'w', encoding='utf-8')
        fw.write(message)
        fw.close()

def txt_title(indexdir, htmldir):
    opentxt = os.path.join( indexdir, "title.txt")
    opentxt = urllib.request.url2pathname(opentxt)
    message = """
  <title>颐亲王的小栈 | 颐亲王 | 正协信息客栈</title>
  <meta name="keywords" content="颐亲王, 郑協, zhengxie.info, 正协, 正协信息" />
  <meta name="description" content="本站点是“颐亲王的小栈”，站点名称为“正协信息客栈”，由“华朝颐亲王郑協”创建并维护，借助“GitHub Pages”，发布于域名“zhengxie.info”内。" />
    """
    if not os.path.exists(opentxt):
        fw = open(opentxt, 'w', encoding='utf-8')
        fw.write(message)
        fw.close()




def txt_footer(txtdir):
    path = openpath(files="footer.txt", dirf=txtdir)
    strtime = time.strftime("%Y-%m-%dT%H:%M:%S", time.gmtime())
    message = """
  <footer><!--足底-->
    <div class="support"><!--基础文档-->      
      &lt;&nbsp;
      <a href="https://zhengxie.info/base/privacy/" title="" target="_self">隐私保护</a>      
      &nbsp;|&nbsp;
      <a href="https://zhengxie.info/base/disclaimer/" title="" target="_self">免责声明</a>
      &nbsp;|&nbsp;
      <a href="https://zhengxie.info/base/copyright/" title="" target="_self">版权协议</a>
      &nbsp;&gt;
    </div>
    <div class="support"><!--版本更新-->      
      &nbsp;|&nbsp;
      &lt;&nbsp;
      版本更新：<lastmod>%s+00:00</lastmod>
      &nbsp;&gt;
      &nbsp;|&nbsp;
    </div>
    <div class="support"><!--版本测试-->
      &lt;&nbsp;
      测试版本：
      <a href="https://zhengxie.info/alpha/" title="" target="_self">alpha</a>
      &nbsp;|&nbsp;
      <a href="https://zhengxie.info/beta/" title="" target="_self">beta</a>
      &nbsp;|&nbsp;
      <a href="https://zhengxie.info/gamma/" title="" target="_self">gamma</a>
      &nbsp;&gt;
    </div>
  </footer>
    """%(strtime)    
    w_message(path=path, message=message)

def openpath(files, dirf=""):
    path = os.path.join(dirf, files)
# #     path = urllib.request.url2pathname(path)
    path = urllib.request.pathname2url(path)
# #     path = os.path.abspath(path)
#     print("地址：", path)
#     if not os.path.exists(path):
#         print("文件", path, "不存在")
    return path

def w_message(path, message):
    if not os.path.exists(path):
        with open(path, 'w', encoding='utf-8') as fw:
            fw.write(message)
        print("创建：", path)

def htmldirs(path):
    list_htmldir = []
    indexdirs = []    
    with open(path, 'r', encoding='utf-8') as fr:
        lines = fr.readlines()
        for line in lines:
            line = line.strip()
            list_htmldir.append(line)
            indexdir = openpath("index", dirf=line)
            if not os.path.exists(indexdir):
                os.mkdir(indexdir)
            indexdirs.append(indexdir)
    return list_htmldir, indexdirs

def r_txt(txtname, txtdir):
    path = openpath(txtname, txtdir)
    if not os.path.exists(path):
        message = "\n"
        print("文件不存在", path)
    else:
        with open(path, 'r', encoding='utf-8') as fr:
            message = fr.read()
    return message

def create_html(htmldir, txtdir, indexdir):
    
    txt_link(indexdir=indexdir, htmldir=htmldir)
    txt_title(indexdir=indexdir, htmldir=htmldir)
    txt_author(indexdir=indexdir, htmldir=htmldir)
    txt_article(indexdir=indexdir, htmldir=htmldir)
    
    path = openpath('index.html', htmldir)
    path = os.path.abspath(path)
    with open(path, 'w', encoding='utf-8') as fw:        
        fw.write(r_txt('head.txt', txtdir))
        fw.write(r_txt('title.txt', indexdir))
        fw.write(r_txt('link.txt', indexdir))
        fw.write(r_txt('author.txt', indexdir))
        fw.write(r_txt('body.txt', txtdir))
        fw.write(r_txt('header.txt', txtdir))
        fw.write('  <div class="central">\n')
        fw.write('    <main>\n')
        fw.write('      <article>\n')
        fw.write(r_txt('article.txt', indexdir))
        fw.write('      </article>\n')
        fw.write('    </main>\n')
        fw.write("  </div>\n")
        fw.write(r_txt('footer.txt', txtdir))
        fw.write(r_txt('script.txt', txtdir))
        fw.write(r_txt('html.txt', txtdir))

#         fw.write("\n</body>\n</html>")
        
    webbrowser.open(path,new = 0, autoraise=True)
    
def execute(pydir):
    txtdir = openpath("../txt")
    txt_footer(txtdir)
#     要创建index.html的目录列表
    path = "htmldir.txt"
    path = openpath(path, txtdir)
    dirlist, indexdirs = htmldirs(path=path)
    
#     运行主程序
    for n in range(len(dirlist)):
        create_html(htmldir=dirlist[n], txtdir=txtdir, indexdir=indexdirs[n])


if __name__ == "__main__":
#     确认初始位置
    pydir = "D:\\Workshop\\python\\seymour-zx.github.io\\common\\python"
    if pydir==os.getcwd():
        execute(pydir=pydir)
    else:
        print("初始位置错误！")
```
