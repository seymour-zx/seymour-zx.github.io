```python
# coding:utf-8
encoding='utf-8'

import os
import urllib.request
import webbrowser

# 记录通用资源的绝对地址
def def_address_resource():
    global address_ico, address_css
    # address_css = os.path.abspath("../css/mystyle.css")
    address_css = os.path.realpath("../css/mystyle.css")
    print('记录css位置:\naddress_css =', address_css)
    address_ico = os.path.realpath("../image/favicon.ico")
    print('记录ico位置:\naddress_ico =', address_ico)

# 切换到创建index.html的区域area_run：
def def_area_run():
    global area_run, input_run
    area_run = input_run
    # area_run = input("切换到运行区域：")
    # if area_run == "":
    #     area_run = os.path.realpath("../../base/test")
    os.chdir(area_run)
    print("完成切换到运行区域：",os.getcwd())
    # 创建的目录
    path = "./index"
    if not os.path.exists(path):
        os.mkdir(path)

# 创建./index/link.txt文件<link>：
def def_create_link():
    global rel_ico, rel_css, address_ico, address_css
    rel_ico = os.path.relpath(address_ico)
    rel_ico = urllib.request.pathname2url(rel_ico)
    rel_css = os.path.relpath(address_css)
    rel_css = urllib.request.pathname2url(rel_css)
    message = """
  <link href="%s" rel="shortcut icon" type="image/x-icon" />
  <!--调用网页图标-->
  <link href="%s" rel="stylesheet" type="text/css" />
  <!--调用css样式-->
    """%(rel_ico, rel_css)
    fw = open('./index/link.txt', 'w', encoding='utf-8')
    fw.write(message)
    fw.close()

# 创建./index/title.txt文件<title>：
def def_create_title():
    message = """
  <title>颐亲王的小栈</title>
  <meta name="author" content="华朝颐亲王郑協" />
  <meta name="keywords" content="颐亲王,郑協,zhengxie.info" />
  <meta name="description" content="本站点是'颐亲王的小栈',站点名称为'正协信息客栈',由'华朝颐亲王郑協'创建并维护,借助'GitHub Pages',发布于域名'zhengxie.info'内。" />
    """
    path = "./index/title.txt"
    if not os.path.exists(path):
        fw = open(path, 'w', encoding='utf-8')
        fw.write(message)
        fw.close()
    else:
        print('位置：',path,' 已存在')

# 创建./index/article.txt文件<article>：
def def_create_article():
    message = """
      <article>
        <h1>模板</h1>
        <p>
            未编辑
                    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
        &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
        &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
        &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
        &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
        &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
        &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
        &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
        &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
        &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
        &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
        </p>
                <hr /><hr /><hr /><hr /><hr /><hr /><hr /><hr /><hr /><hr />
        <hr /><hr /><hr /><hr /><hr /><hr /><hr /><hr /><hr /><hr />
        <hr /><hr /><hr /><hr /><hr /><hr /><hr /><hr /><hr /><hr />

        <hr /><hr /><hr /><hr /><hr /><hr /><hr /><hr /><hr /><hr />
        <hr /><hr /><hr /><hr /><hr /><hr /><hr /><hr /><hr /><hr />
        <hr /><hr /><hr /><hr /><hr /><hr /><hr /><hr /><hr /><hr />
        <hr /><hr /><hr /><hr /><hr /><hr /><hr /><hr /><hr /><hr />
        <p>
            未编辑
                    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
        &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
        &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
        &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
        &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
        &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
        &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
        &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
        &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
        &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
        &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
        </p>

      </article>

    """
    path = "./index/article.txt"
    if not os.path.exists(path):
        fw = open(path, 'w', encoding='utf-8')
        fw.write(message)
        fw.close()
    else:
        print('位置：',path,' 已存在')


# 创建index.html
def def_create_html():
    global address_ipynb, area_run, input_run

    # 记录初始位置，为了防止再次运行时，位置出错，最后要返回初始位置。
    address_ipynb = "D:\\Workshop\\html\\seymour-zx.github.io\\common\\python"
    print('记录ipynb位置:\naddress_ipynb =', address_ipynb)

    # 记录通用资源的绝对地址
    def_address_resource()

    # 切换到创建index.html的区域area_run：
    def_area_run()

    # 创建./index/link.txt文件<link>：
    def_create_link()

    # 创建./index/title.txt文件<title>：
    def_create_title()

    # 创建./index/article.txt文件<article>：
    def_create_article()

    # 合并排序
    txtname = []
    txtpath = []
    txt = []
    txtpath.append(address_ipynb)
    txtname.append("../txt/head.txt")
    txtpath.append(area_run)
    txtname.append("./index/link.txt")
    txtpath.append(area_run)
    txtname.append("./index/title.txt")
    txtpath.append(address_ipynb)
    txtname.append("../txt/body.txt")
    txtpath.append(area_run)
    txtname.append("./index/article.txt")
    txtpath.append(address_ipynb)
    txtname.append("../txt/script.txt")


    html = "./index.html"
    os.chdir(area_run)
    html = os.path.realpath(html)
    fw = open(html, 'w', encoding='utf-8')
    for i in range(6):
        os.chdir(txtpath[i])
        path = os.path.realpath(txtname[i])
        txt.append(path)
        if not os.path.exists(txt[i]):
            with open(txt[i], 'w', encoding='utf-8') as fr:
                fr.close()
                print("文件",txt[i],"不存在")
        else:
            with open(txt[i], 'r', encoding='utf-8') as fr:
                fw.write('\n')
                for j in fr:
                    fw.write(j)
                fr.close()
    fw.close()
    webbrowser.open(html,new = 0, autoraise=True)


    # 返回初始位置，是为了防止再次运行时，位置出错。
    os.chdir(address_ipynb)
    print('变更回初始运行区域:', os.getcwd())

# 主程序
creathtml = []
creathtml.append("../../base/test")
creathtml.append("../../base/about")
creathtml.append("../../base/changelog")
creathtml.append("../../base/copyright")
creathtml.append("../../base/disclaimer")
creathtml.append("../../base/homepage")
creathtml.append("../../base/privacy")
creathtml.append("../../base/sitemap")
creathtml.append("../../private/bookmark")

print("将要创建 ",len(creathtml)," 个index.html")
for n in range(len(creathtml)):
    input_run = creathtml[n]
    def_create_html()
    print(n+1,"完成")
```
